```python
from qiskit import *
qr = QuantumRegister(5)
cr = ClassicalRegister(5)
xxxxa = Quantumcircuit(qr, cr)
%matplotlib inline
xxxxa.h(qr[0])
xxxxa.h(qr[2])
xxxxa.h(qr[3])
xxxxa.cx(qr[0],qr[1])
xxxxa.h(qr[0])
xxxxa.cx(qr[2],qr[1])
xxxxa.h(qr[2])
xxxxa.cx(qr[3],qr[1])
xxxxa.h(qr[3])
xxxxa.h(qr[1])

xxxxa.barrier(qr)

xxxxa.h(qr[0])
xxxxa.h(qr[1])
xxxxa.h(qr[2])
xxxxa.tdg(qr[3])
xxxxa.h(qr[3])
xxxxa.measure(qr, cr)
xxxxa.draw(output='mpl')
```

```python
IBMQ.load_account()
provider = IBMQ.get_provider('ibm-q')
qcomp = provider.get_backend('ibmq_athens')
job = execute(xxxxa, backend=qcomp,shots = 8192)
from qiskit.tools.monitor import job_monitor
```

```python
job_monitor(job)
```

```python
result = job.result()
from qiskit.tools.visualization import plot_histogram
plot_histogram(result.get_counts(circuit))
```
