```python
from qiskit import *
qr1 = QuantumRegister(5)
cr1 = ClassicalRegister(5)
xxxxa = QuantumCircuit(qr1, cr1)
%matplotlib inline
xxxxa.h(qr1[0])
xxxxa.h(qr1[1])
xxxxa.h(qr1[3])
xxxxa.h(qr1[4])
xxxxa.cx(qr1[1],qr1[2])
xxxxa.h(qr1[1])
xxxxa.cx(qr1[3],qr1[2])
xxxxa.h(qr1[3])
xxxxa.cx(qr1[4],qr1[2])
xxxxa.cx(qr1[0],qr1[2])
xxxxa.h(qr1[0])
xxxxa.h(qr1[2])
xxxxa.h(qr1[4])

xxxxa.barrier(qr1)

xxxxa.h(qr1[0])
xxxxa.h(qr1[1])
xxxxa.h(qr1[2])
xxxxa.h(qr1[3])
xxxxa.tdg(qr1[4])
xxxxa.h(qr1[4])
xxxxa.measure(qr1, cr1)
xxxxa.draw(output='mpl')
```
