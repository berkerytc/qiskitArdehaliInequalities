```python
from qiskit import *
qr1 = QuantumRegister(5)
cr1 = ClassicalRegister(5)
yyyya = QuantumCircuit(qr1, cr1)
%matplotlib inline
yyyya.h(qr1[0])
yyyya.h(qr1[1])
yyyya.h(qr1[3])
yyyya.h(qr1[4])
yyyya.cx(qr1[1],qr1[2])
yyyya.h(qr1[1])
yyyya.cx(qr1[3],qr1[2])
yyyya.h(qr1[3])
yyyya.cx(qr1[4],qr1[2])
yyyya.cx(qr1[0],qr1[2])
yyyya.h(qr1[0])
yyyya.h(qr1[2])
yyyya.h(qr1[4])

yyyya.barrier(qr1)

yyyya.sdg(qr1[0])
yyyya.h(qr1[0])
yyyya.sdg(qr1[1])
yyyya.h(qr1[1])
yyyya.sdg(qr1[2])
yyyya.h(qr1[2])
yyyya.sdg(qr1[3])
yyyya.h(qr1[3])
yyyya.tdg(qr1[4])
yyyya.h(qr1[4])
yyyya.measure(qr1, cr1)
yyyya.draw(output='mpl')
```

```python
qr2 = QuantumRegister(5)
cr2 = ClassicalRegister(5)
yyyyb = QuantumCircuit(qr2, cr2)
%matplotlib inline
yyyyb.h(qr2[0])
yyyyb.h(qr2[1])
yyyyb.h(qr2[3])
yyyyb.h(qr2[4])
yyyyb.cx(qr2[1],qr2[2])
yyyyb.h(qr2[1])
yyyyb.cx(qr2[3],qr2[2])
yyyyb.h(qr2[3])
yyyyb.cx(qr2[4],qr2[2])
yyyyb.cx(qr2[0],qr2[2])
yyyyb.h(qr2[0])
yyyyb.h(qr2[2])
yyyyb.h(qr2[4])

yyyyb.barrier(qr2)

yyyyb.sdg(qr2[0])
yyyyb.h(qr2[0])
yyyyb.sdg(qr2[1])
yyyyb.h(qr2[1])
yyyyb.sdg(qr2[2])
yyyyb.h(qr2[2])
yyyyb.sdg(qr2[3])
yyyyb.h(qr2[3])
yyyyb.t(qr2[4])
yyyyb.h(qr2[4])
yyyyb.measure(qr2, cr2)
yyyyb.draw(output='mpl')
```

```python
IBMQ.load_account()
provider = IBMQ.get_provider('ibm-q')
qcomp = provider.get_backend('ibmq_athens')
job1 = execute(yyyya, backend=qcomp,shots = 8192)
job2 = execute(yyyyb, backend=qcomp,shots = 8192)
from qiskit.tools.monitor import job_monitor
```

```python
job_monitor(job1)
job_monitor(job2)
```
