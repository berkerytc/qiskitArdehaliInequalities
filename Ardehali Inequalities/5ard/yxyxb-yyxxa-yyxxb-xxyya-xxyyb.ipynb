```python
from qiskit import *
qr1 = QuantumRegister(5)
cr1 = ClassicalRegister(5)
yxyxb = QuantumCircuit(qr1, cr1)
%matplotlib inline
yxyxb.h(qr1[0])
yxyxb.h(qr1[1])
yxyxb.h(qr1[3])
yxyxb.h(qr1[4])
yxyxb.cx(qr1[1],qr1[2])
yxyxb.h(qr1[1])
yxyxb.cx(qr1[3],qr1[2])
yxyxb.h(qr1[3])
yxyxb.cx(qr1[4],qr1[2])
yxyxb.cx(qr1[0],qr1[2])
yxyxb.h(qr1[0])
yxyxb.h(qr1[2])
yxyxb.h(qr1[4])

yxyxb.barrier(qr1)

yxyxb.sdg(qr1[0])
yxyxb.h(qr1[0])
yxyxb.h(qr1[1])
yxyxb.sdg(qr1[2])
yxyxb.h(qr1[2])
yxyxb.h(qr1[3])
yxyxb.t(qr1[4])
yxyxb.h(qr1[4])
yxyxb.measure(qr1, cr1)
yxyxb.draw(output='mpl')
```
