```python
from qiskit import *
qr1 = QuantumRegister(4)
cr1 = ClassicalRegister(4)
xyxa = QuantumCircuit(qr1, cr1)
%matplotlib inline
xyxa.h(qr1[0])
xyxa.h(qr1[2])
xyxa.h(qr1[3])
xyxa.cx(qr1[0],qr1[1])
xyxa.h(qr1[0])
xyxa.cx(qr1[2],qr1[1])
xyxa.h(qr1[2])
xyxa.cx(qr1[3],qr1[1])
xyxa.h(qr1[3])
xyxa.h(qr1[1])

xyxa.barrier(qr1)

xyxa.h(qr1[0])
xyxa.sdg(qr1[1])
xyxa.h(qr1[1])
xyxa.h(qr1[2])
xyxa.tdg(qr1[3])
xyxa.h(qr1[3])
xyxa.measure(qr1, cr1)
xyxa.draw(output='mpl')
```

```python
job_monitor(job1)
job_monitor(job2)
job_monitor(job3)
job_monitor(job4)
job_monitor(job5)
```
