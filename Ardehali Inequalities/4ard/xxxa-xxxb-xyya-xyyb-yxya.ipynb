```python
from qiskit import *
qr1 = QuantumRegister(4)
cr1 = ClassicalRegister(4)
xxxa = QuantumCircuit(qr1, cr1)
%matplotlib inline
xxxa.h(qr1[0])
xxxa.h(qr1[2])
xxxa.h(qr1[3])
xxxa.cx(qr1[0],qr1[1])
xxxa.h(qr1[0])
xxxa.cx(qr1[2],qr1[1])
xxxa.h(qr1[2])
xxxa.cx(qr1[3],qr1[1])
xxxa.h(qr1[3])
xxxa.h(qr1[1])

xxxa.barrier(qr1)

xxxa.h(qr1[0])
xxxa.h(qr1[1])
xxxa.h(qr1[2])
xxxa.tdg(qr1[3])
xxxa.h(qr1[3])
xxxa.measure(qr1, cr1)
xxxa.draw(output='mpl')
```
