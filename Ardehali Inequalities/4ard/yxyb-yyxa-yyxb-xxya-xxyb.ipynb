```python
from qiskit import *
qr1 = QuantumRegister(4)
cr1 = ClassicalRegister(4)
yxyb = QuantumCircuit(qr1, cr1)
%matplotlib inline
yxyb.h(qr1[0])
yxyb.h(qr1[2])
yxyb.h(qr1[3])
yxyb.cx(qr1[0],qr1[1])
yxyb.h(qr1[0])
yxyb.cx(qr1[2],qr1[1])
yxyb.h(qr1[2])
yxyb.cx(qr1[3],qr1[1])
yxyb.h(qr1[3])
yxyb.h(qr1[1])

yxyb.barrier(qr1)

yxyb.sdg(qr1[0])
yxyb.h(qr1[0])
yxyb.h(qr1[1])
yxyb.sdg(qr1[2])
yxyb.h(qr1[2])
yxyb.t(qr1[3])
yxyb.h(qr1[3])
yxyb.measure(qr1, cr1)
yxyb.draw(output='mpl')
```
